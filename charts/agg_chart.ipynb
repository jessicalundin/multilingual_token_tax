```python
df = pd.read_csv('full/accuracy_by_subject_language.csv')
# # Drop o1-preview column
# df = df.drop(columns=['OpenAI/o1-preview-2024-09-12'])
# Drop any rows with language 'fra' and 'eng'
df = df[df['language'] != 'eng']
df = df[df['language'] != 'fra']
# Drop the 'language' column before grouping
df = df.drop(columns=['language'])
# Group by subject and average across all languages
df = df.groupby('subject').mean().reset_index()
# Swap rows and columns
df = df.set_index('subject').T
# Rename the index to 'model'
df.index.name = 'model'
# Rename the columns to 'subject'
df.columns.name = 'subject'
# Add a new column 'total' that is the average of all columns
df['total'] = df.mean(axis=1)
# Reorder the columns to have 'total' at the end
df = df[['total'] + [col for col in df.columns if col != 'total']]
# round the values to 3 decimal places
df = df.round(3)
# multiply by 100
df = df * 100
# Reorder rows to match the model_order
df = df.reindex(model_order)
# Change row names to be more readable (keep only everything after the last '/')
df.index = df.index.str.split('/').str[-1]
df
```

```python
model_df = pd.read_csv('../../model_df.csv')
# Drop 'OpenAI_o1'
model_df = model_df[model_df['dir'] != 'OpenAI_o1']
sums = []
for dir in model_df['dir']:
    # Read the CSV file
    df = pd.read_csv(f'full/{dir}_full_results.csv')
    # Sum the 'num_tokens' column
    sum = df['num_tokens'].sum()
    # Append the sum to the list
    sums.append(sum)
# Add the sums as a new column to the model_df DataFrame
model_df['num_tokens'] = sums
# Drop all columns except 'dir', 'tokenizer', and 'num_tokens'
model_df = model_df[['dir', 'tokenizer', 'num_tokens']]
# Add an avg row
avg_row = pd.DataFrame({'dir': ['Avg'], 'num_tokens': [model_df['num_tokens'].mean()]})
# Add a total row (without the avg)
total_row = pd.DataFrame({'dir': ['Total'], 'num_tokens': [model_df['num_tokens'].sum()]})
# Append the avg and total rows to the model_df DataFrame
model_df = pd.concat([model_df, avg_row], ignore_index=True)
model_df = pd.concat([model_df, total_row], ignore_index=True)
# Rename the 'dir' column to 'Model'
model_df = model_df.rename(columns={'dir': 'Model'})
# Set the display option to show num_tokens as int
pd.set_option('display.float_format', '{:.0f}'.format)

model_df
```
