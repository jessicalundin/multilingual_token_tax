```python
import pandas as pd
import glob

# 1. Read & combine all your result CSVs
all_files = glob.glob('*full_results.csv')
dfs = []
for fn in all_files:
    df = pd.read_csv(fn)
    # make sure fertility is numeric
    df['fertility_row_wise'] = df['fertility_row_wise'].astype(float)
    dfs.append(df)
df_all = pd.concat(dfs, ignore_index=True)

# 2. Prepare lists
languages = sorted(df_all['language'].unique())
subjects  = sorted(df_all['subject'].unique())

# 3. Loop through each language
for lang in languages:
    df_lang = df_all[df_all['language'] == lang]
    
    # 3a. Total fertility per model
    total_fert = df_lang.groupby('llm')['fertility_row_wise'].mean()
    total_df   = total_fert.to_frame().T
    total_df.index = [f"Total {lang}"]
    
    # 3b. Per‐subject fertility per model
    subj_fert = (
        df_lang
        .pivot_table(
            index='subject',
            columns='llm',
            values='fertility_row_wise',
            aggfunc='mean',
            fill_value=0
        )
    )
    
    # 3c. Combine rows, put Total first, keep same subject order
    combined = pd.concat([total_df, subj_fert])
    combined = combined.reindex([f"Total {lang}"] + subjects)
    
    # 3d. Alphabetical model columns
    combined = combined[sorted(combined.columns)]
    
    # 3e. Print Markdown table
    print(f"\n### Fertility (row-wise) for {lang}\n")
    print(combined.round(2).to_markdown())
```
