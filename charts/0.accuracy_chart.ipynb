```python
import pandas as pd

# 1. Load your file
df = pd.read_csv('accuracy_Llama31_405B.csv')

# 2. Ensure accuracy is boolean
df['accuracy'] = df['accuracy'].astype(bool)

# 3. Grab the model name (just to label the column)
model_name = df['llm'].iloc[0]  # should be "Llama31_405B"

# 4. Precompute the sorted list of subjects (for consistent ordering)
subjects = sorted(df['subject'].unique())

# 5. Loop over each language
for lang in sorted(df['language'].unique()):
    df_lang = df[df['language'] == lang]

    # 5a. Compute per‐subject % correct
    subj_acc = df_lang.groupby('subject')['accuracy'].mean() * 100

    # 5b. Build a DataFrame and name the column after the model
    table = subj_acc.to_frame(name=model_name)

    # 5c. Compute the overall % correct for this language
    total_acc = df_lang['accuracy'].mean() * 100
    table.loc[f"Total {lang}"] = total_acc

    # 5d. Reorder so the Total row sits at the top
    table = table.reindex([f"Total {lang}"] + subjects)

    # 5e. Print out a nice Markdown table
    print(f"\n### Accuracy Breakdown for {lang} ({model_name})\n")
    print(table.round(2).to_markdown())
```

```python
import pandas as pd
import glob

# 1. Read & combine all model CSVs
all_files = glob.glob('accuracy_*.csv')
dfs = []
for fn in all_files:
    df = pd.read_csv(fn)
    df['accuracy'] = df['accuracy'].astype(bool)
    dfs.append(df)
df_all = pd.concat(dfs, ignore_index=True)

# 2. Prepare lists
languages = sorted(df_all['language'].unique())
subjects  = sorted(df_all['subject'].unique())

# 3. Loop through each language
for lang in languages:
    df_lang = df_all[df_all['language'] == lang]
    
    # 3a. Total accuracy per model
    total_acc = df_lang.groupby('llm')['accuracy'].mean() * 100
    total_df = pd.DataFrame(total_acc).T
    total_df.index = [f"Total {lang}"]
    
    # 3b. Per‐subject accuracy per model
    subj_pct = (df_lang
                .pivot_table(
                    index='subject',
                    columns='llm',
                    values='accuracy',
                    aggfunc='mean',
                    fill_value=0
                ) * 100
               )
    
    # 3c. Combine rows and reorder so “Total” is first
    combined = pd.concat([total_df, subj_pct])
    combined = combined.reindex([f"Total {lang}"] + subjects)
    
    # 3d. **Alphabetical** column order
    combined = combined[sorted(combined.columns)]
    
    # 3e. Print Markdown table
    print(f"\n### Accuracy Breakdown for {lang}\n")
    print(combined.round(2).to_markdown())
```

```python
import pandas as pd
import glob

# 1. Read & combine all model CSVs
all_files = glob.glob('accuracy_*.csv')
dfs = []
for fn in all_files:
    df = pd.read_csv(fn)
    df['accuracy'] = df['accuracy'].astype(bool)
    dfs.append(df)
df_all = pd.concat(dfs, ignore_index=True)

# 2. Prepare lists
languages = sorted(df_all['language'].unique())
subjects  = sorted(df_all['subject'].unique())

# 3. Loop per language, build table, and emit LaTeX
for lang in languages:
    df_lang = df_all[df_all['language'] == lang]

    # 3a. Per‐model overall %
    total = (df_lang.groupby('llm')['accuracy']
                  .mean() * 100)
    total_df = pd.DataFrame(total).T
    total_df.index = [f"Total {lang}"]

    # 3b. Per‐subject %
    subj_df = (df_lang.pivot_table(
                    index='subject',
                    columns='llm',
                    values='accuracy',
                    aggfunc='mean',
                    fill_value=0
               ) * 100)

    # 3c. Combine & reorder rows
    table = pd.concat([total_df, subj_df])
    table = table.reindex([f"Total {lang}"] + subjects)

    # 3d. Alphabetical model columns
    table = table[sorted(table.columns)]

    # 3e. Emit LaTeX
    print(table
        .round(2)
        .to_latex(
            index=True,
            caption=f"Accuracy Breakdown for {lang}",
            label=f"tab:accuracy_{lang}",
            float_format="%.2f",
            column_format="l" + "r" * len(table.columns)
        )
    )
```
